- 정규분포를 따르는 데이터(x)를 200개 생성하고 (평균 15, 표준편차 3),

- x의 zscore (xZscores)를 계산하고,

- 그래프 x는 행1열1, xZscores는 행1열2의 위치에 2개를 출력하시오.

```python
import numpy as np
from scipy import stats


# 정규분포를 따르는 데이터(x)를 200개 생성하고 (평균 15, 표준편차 3),
x = np.random.normal(15, 3, size=200)
x
```

```python
# x의 zscore (xZscores)를 계산
xZscores = stats.zscore(x)
xZscores
```

```python
import matplotlib.pyplot as plt

# Figure 생성
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # 1행 2열

# 첫 번째 subplot ->  x 그래프 
ax[0].plot(x, label='x')
ax[0].set_title('x')

# 두 번째 subplot -> xZscores 그래프 
ax[1].plot(xZscores, label='xZscores')
ax[1].set_title('xZscores')


# 그래프 표시
plt.show()
```
