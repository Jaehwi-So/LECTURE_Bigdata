### 2023년 10월 1일 7016번 버스에 대해 승하차인원을 수집하여 다음과 같이 일자, 승차인원, 하차인원, 정류장명을 100건 출력하자.

맨 뒤에 승차인원 합계를 출력한다.


### 1. Secret Key 로드

```python
import os
import src.mylib as mylib

keyPath = os.path.join(os.getcwd(), 'src', 'key.properties')
key = mylib.getKey(keyPath)
key
```

### 2. API URL 및 파라미터 설정

```python
_url='http://openAPI.seoul.go.kr:8088'
_key=str(key['dataseoul'])
_type='json'
_service='CardBusStatisticsServiceNew'
_start_index=1
_end_index=100
_use_dt='20231001'
_bus_route_no='7016'
```

```python
import urllib

_api=os.path.join(_url,_key,_type,_service,str(_start_index),str(_end_index),_use_dt,_bus_route_no)
                           
print("api: ", _api)
```

### 3. API 요청 및 JSON Parsing

```python
import requests
response = requests.get(_api)

print(response.text)
```

```python
json = response.json()
json
```

### 4. 수집한 데이터 출력

```python
rows = json['CardBusStatisticsServiceNew']['row']
sum_ride_pasgr_num = 0;
for row in rows:
    print ("{0:5s}\t{1:10.1f}\t{2:9.1f}\t{3:9s}"
           .format(row['USE_DT'],row['RIDE_PASGR_NUM'],row['ALIGHT_PASGR_NUM'],row['BUS_STA_NM']))
    sum_ride_pasgr_num += row['RIDE_PASGR_NUM']
    
print(f'승차인원 합계: {sum_ride_pasgr_num}')
```
